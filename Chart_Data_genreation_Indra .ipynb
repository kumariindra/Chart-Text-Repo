# Dependencies

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import os
import random
%matplotlib inline
```

# Reading the Chart Titles

```python
chart_titles=pd.read_csv('chart_titles.csv')
```

```python
chart_titles=np.array(chart_titles.values).reshape(-1)
```

# Setting Charts Width and Height

```python
width = 500 / 72.0
height = 500 / 72.0
```

```python
y=np.random.randint(0,1000,size=(200,500))
```

```python
import nltk
from nltk.corpus import words
import random
nltk.download('words')
word_list = words.words()
# generate 2000 random words
random_words = [random.choice(word_list) for i in range(2000)]
#filtering those words which have length of 10 or less
random_words = [word for word in random_words if len(word)<=10]
#print(random_words)
```

```python
x=np.array(random_words)
```

# **1. Vertical Bar Charts**

```python
if not os.path.exists("1. vertical_bar_charts"):
    os.makedirs("1. vertical_bar_charts")
for i in range(10):
  # Select 5 random values from the array
  x_values =x.flatten() [np.random.choice(x.size, 5, replace=False)]
  y_values = y.flatten()[np.random.choice(y.size, 5, replace=False)]
  title_values=chart_titles[np.random.choice(chart_titles.size, 3, replace=False)]
  plt.figure(figsize=(width,height))
  sns.barplot(x=x_values,y=y_values)
  plt.title(title_values[0])
  plt.xlabel(title_values[1],labelpad = 20)
  plt.ylabel(title_values[2],labelpad = 20)
  plt.savefig(f'./1. vertical_bar_charts/Bar Chart {i+1}.png')
  plt.show()
  plt.close()
```

# **2. Bar Chart with Rotated Labels**

```python
if not os.path.exists("2. vertical_bar_charts_with rotation"):
    os.makedirs("2. vertical_bar_charts_with rotation")
for i in range(1):
  # Select 5 random values from the array
  x_values =x.flatten() [np.random.choice(x.size, 5, replace=False)]
  y_values = y.flatten()[np.random.choice(y.size, 5, replace=False)]
  title_values=chart_titles[np.random.choice(chart_titles.size, 3, replace=False)]
  plt.figure(figsize=(width,height))
  sns.barplot(x=x_values,y=y_values)
  plt.xticks(rotation=45)
  plt.title(title_values[0])
  plt.xlabel(title_values[1],labelpad = 10)
  plt.ylabel(title_values[2],labelpad = 10)
  plt.savefig(f'./2. vertical_bar_charts_with rotation/Bar Chart with rotation {i+1}.png')
  plt.show()
```

# **3. Vertical Stacked Bar Charts**

```python
if not os.path.exists("3. vertical_stacked_bar_charts"):
    os.makedirs("3. vertical_stacked_bar_charts")
for i in range(1):
  labels=x.flatten()[np.random.choice(x.size, 3, replace=False)]
  data = {'Category':labels, 
          labels[0]: np.random.choice(y.size, 3, replace=False), 
          labels[1]: np.random.choice(y.size, 3, replace=False),
          labels[2]: np.random.choice(y.size, 3, replace=False)}
  df = pd.DataFrame(data)
  plt.figure(figsize=(width,height))
  df.plot(kind='bar', stacked=True, color=['red', 'skyblue', 'green'])
  title_values=chart_titles[np.random.choice(chart_titles.size, 3, replace=False)]
  plt.title(title_values[0])
  plt.xlabel(title_values[1],labelpad = 10)
  plt.ylabel(title_values[2],labelpad = 10)
  
  plt.savefig(f'./3. vertical_stacked_bar_charts/Stacked Bar Chart{i+1}.png')
  plt.show()
```

# **4. Horizontal Bar Charts**

```python
if not os.path.exists("4. horiznontal_bar_charts"):
    os.makedirs("4. horiznontal_bar_charts")
for i in range(1):
  # Select 5 random values from the array
  x_values =x.flatten() [np.random.choice(x.size, 5, replace=False)]
  y_values = y.flatten()[np.random.choice(y.size, 5, replace=False)]
  title_values=chart_titles[np.random.choice(chart_titles.size, 3, replace=False)]
  plt.figure(figsize=(width,height))
  sns.barplot(y=x_values,x=y_values)
  plt.title(title_values[0])
  plt.xlabel(title_values[1],labelpad = 10)
  plt.ylabel(title_values[2],labelpad = 10)
  plt.savefig(f'./4. horiznontal_bar_charts/Horizontal Bar Chart{i+1}.png')
  plt.show()
```

# **5. Horizontal Stacked Bar Chart**

```python
if not os.path.exists("5. horizontal_stacked_bar_charts"):
    os.makedirs("5. horizontal_stacked_bar_charts")
for i in range(1):
  labels=x.flatten()[np.random.choice(x.size, 3, replace=False)]
  data = {'Category':labels, 
          labels[0]: np.random.choice(y.size, 3, replace=False), 
          labels[1]: np.random.choice(y.size, 3, replace=False),
          labels[2]: np.random.choice(y.size, 3, replace=False)}
  df = pd.DataFrame(data)
  plt.figure(figsize=(width,height))
  df.plot.barh(stacked=True, color=['red', 'skyblue', 'green'])
  title_values=chart_titles[np.random.choice(chart_titles.size, 3, replace=False)]
  plt.title(title_values[0])
  plt.xlabel(title_values[1],labelpad = 10)
  plt.ylabel(title_values[2],labelpad = 10)
  plt.savefig(f'./5. horizontal_stacked_bar_charts/Horizontal Stacked Bar Chart{i+1}.png')
  plt.show()
```

# **6. Line Charts**

```python
if not os.path.exists("6. line_charts"):
    os.makedirs("6. line_charts")
for i in range(1):
  # Select 5 random values from the array
  x_values =x.flatten() [np.random.choice(x.size, 5, replace=False)]
  y_values = y.flatten()[np.random.choice(y.size, 5, replace=False)]
  title_values=chart_titles[np.random.choice(chart_titles.size, 3, replace=False)]
  plt.figure(figsize=(width,height))
  sns.lineplot(x=x_values,y=y_values)
  plt.title(title_values[0])
  plt.xlabel(title_values[1],labelpad = 10)
  plt.ylabel(title_values[2],labelpad = 10)
  plt.savefig(f'./6. line_charts/Line Chart{i+1}.png')
  plt.show()
```

# **7. PIE Chart**

```python
if not os.path.exists("7. pie_charts"):
    os.makedirs("7. pie_charts")
for i in range(2):
  plt.figure(figsize=(width,height))
  x_values =x.flatten() [np.random.choice(x.size, 5, replace=False)]
  y_values = y.flatten()[np.random.choice(y.size, 5, replace=False)]
  title_values=chart_titles[np.random.choice(chart_titles.size, 1, replace=False)]
  plt.pie(y_values, labels = x_values)
  plt.title(title_values[0])

  plt.savefig(f'./7. pie_charts/Pie Chart{i+1}.png')
  plt.show()
```
